# Using Kelly formula to decide the amount of betting

https://bjlkeng.github.io/posts/optimal-betting-and-the-kelly-criterion/ <br>

https://en.wikipedia.org/wiki/Kelly_criterion

$f^{*} = \frac{p}{a}- \frac{q}{b}$

p : 勝率 <br>
q = 1-p <br>
a = the fraction that is lost in a negative outcome <br>
b = the fraction that is gained in a positive outcome

參數p 參考alpha_030_S&P500_Methodology檔案提及之 P(increase | alpha030 >= 0.125) <br>
q = 1-p

a = 策略交易平均虧損(只看虧損)

b = 策略交易平均獲利(只看獲利)

動態調整下注比例： 先假設第一次交易的獲益率為前一年平均獲益率，根據歷史的機率之後計算出第一次下注的比例。 之後得到第一次交易的實際報酬，調整後的發生機率，計算出下一次的下注比例。

```python
p_de = 125 #參考前一份檔案之sum(multiply['AAPL']['2021'] == 1)
p_nu = 67  #參考前一份檔案之intersect_alpha_030
```

```python
p = p_nu/p_de
q = 1-p
a = abs(sum(neg_ret)/len(neg_ret))
b = sum(posi_ret)/len(posi_ret)
```

```python
f = p/a - q/b
```

```python
f
```

```python
initial_capital = 1000000
daily_end_capital = initial_capital
daily_begin_capital = initial_capital
long_pos = 0
buy_price = 0
sell_price = 0
daily_ret = 0
earn = []
ret = []
posi_ret_2 = []
neg_ret_2 = []
posi_pre = posi_ret
neg_pre = neg_ret
var = []
risk_exposure = []
```

```python
date_2022 = c['AAPL']['2022'].index.format(formatter=lambda x: x.strftime('%Y%m%d%H:%M.%S'))
```

```python
alpha_date = alpha_030['AAPL']['2022'].index.format(formatter=lambda x: x.strftime('%Y%m%d%H:%M.%S'))
c_date = c['AAPL']['2022'].index.format(formatter=lambda x: x.strftime('%Y%m%d%H:%M.%S'))
```

```python
for i in range(1,len(date_2022)-2):
    if daily_begin_capital<=0:
        print(f"在 {date_2022[i]} 破產，結束交易")
        break
    if alpha_030['AAPL']['2022'][i] >=0.125:
        buy_price = c['AAPL']['2022'][i]
        long_pos = round(f*(1/100) * daily_begin_capital/buy_price) # 假設
        daily_end_capital -= buy_price * long_pos #以 今天的收盤價 買進(買不到過去的股票)
        print(f"在 {date_2022[i]} 根據第1個策略，以 {buy_price}買進")
        if daily_end_capital <100000:
            print('注意：在買進之後之後，資本小於10%')
        p_de +=1
        sell_price = c['AAPL']['2022'][i+1]
        daily_ret = (sell_price - buy_price)/buy_price
        if daily_ret >=0:
            p_nu+=1
            posi_pre.append(daily_ret)
            posi_pre.pop(0)
            posi_ret_2.append((date_2022[i],daily_ret))
        else:
            neg_pre.append(daily_ret)
            neg_pre.pop(0)
            neg_ret_2.append((date_2022[i],daily_ret))
        ret.append(daily_ret)
        earn.append(sell_price * long_pos)
        print(f"在 {date_2022[i+1]} 根據第1個策略，以 {sell_price}賣出，單次交易報酬：{daily_ret}")
        daily_end_capital += sell_price * long_pos
        a = abs(sum(neg_pre)/len(neg_pre))
        b = sum(posi_pre)/len(posi_pre)
        p = p_nu/p_de
        q = 1-p
        f = p/a - q/b
        LGD = sum(neg_pre)/len(neg_pre)
        risk = q * LGD
        risk_exposure.append((date_2022[i],risk))
        value = np.percentile(np.array(ret),5)
        var.append(value)
        if daily_ret < value:
            print(f"在 {date_2022[i+1]} 報酬率小於VaR")
    daily_begin_capital = daily_end_capital
total_ret = daily_end_capital/initial_capital
```

# Risk Exposure

定義： 一筆交易的暴險值為 <br>
$\text{Probability of Default} * \text{Loss Given Default}$ <br>
但是違約率不好估計，我先以發生損失的機率跟損失多少當作計算依據。

PD = q <br>
LGD = neg_pre.mean()
這樣計算出來的risk exposure為 期望損失比例

1. 獲利面：<br>
總計78筆交易中，38筆獲利，40筆虧損。<br>
報酬率為-0.004，表現不佳。<br>

在以下日期的獲利大於3%：<br>
0127,0308,0427,0512,0726,0802<br>
整體表現不佳。<br>
<br>
2. 損失面： <br>
最大虧損：-0.05641911877781425 <br>
在以下日期的虧損大於3%： <br>
0428, 0517 <br>
<br>
突破VaR的日期：<br>
0310,0422,0429,0518 <br>
<br>
虧損高於3%之交易，衡量是否在之後幾天交易較佳之下，需要持有的天數過長，違反此策略之精神，故應該選擇不做此交易。<br>
<br>
3. 風險面：<br>
假定風險承受程度為：單筆虧損不超過10% <br>
此交易架構下，目前沒有出現超出風險承受程度的交易。<br>
<br>
衡量Sharpe Ratio：Sharpe ratio最低為-7 最高為8 平均<0 中位數<0，表示此交易策略在承擔風險之下，表現比0.01%還要差，顯示獲利的能力不足。<br>
<br>
衡量MDD：MDD 平均為-4%，表示股價的完整跌幅平均而言落在這個數字。因為此筆交易持有期間不長，持有部位的價值不太因為MDD而受影響。<br>
<br>
VaR：因為交易虧損的數量比較多，使得VaR衡量出來為負值。<br>
報酬往下穿透VaR的日期為：0310,0422,0429,0518 ，其中0429與0518為虧損超過3%之交易，需要特別注意報酬穿透VaR隻交易情形。<br>
risk exposure：暴險值算法此處為價格下跌機率 * 平均虧損比例。衡量出來的值為：在價格下跌之下，平均而言會受到的損失為多少。<br>
整體而言暴險並不高，儘管因為算法導致整體的暴險值看起來很小，但因為價格下降的機率差不多為0.4左右，儘管單看平均損失比例也不會太高。<br>
<br>
暴險的來源：持有天期短，暴險來自於短期的價格波動所影響到的決策。其餘方面的風險比較小。<br>
<br>
衡量是否繼續採行此策略：整體獲利不理想，Sharpe ratio顯示儘管承受風險，獲得的報酬平均而言小於0.01%(設定的無風險利率)，交易獲利與虧損的情況大致上為1:1，但獲利並不能填平虧損。風險都落在能承受的範圍，若能提升交易獲利的勝率，此交易應該繼續進行，若持續後發現仍維持現階段之情況，或許可以考慮別的策略。
