# 策略

多頭部位持有期間: 1天。 <br>
alpha_030 > 0.125就進場

Strategy：When alpha_030 >= 0.125就進場，隔天出場

```python
def strategy_1(names,year,holding_period):
    """
    name : 股票代碼
    year : 用字串
    """
    date = c[names[0]][year].index.format(formatter=lambda x: x.strftime('%Y %m %d %H:%M.%S'))
    initial_capital = 1000000
    daily_end_capital = initial_capital
    daily_begin_capital = initial_capital
    long_pos = 0
    buy_price = 0
    sell_price = 0
    daily_ret = 0
    earn = []
    ret = {}
    posi_ret = {}
    neg_ret = {}
    for i in range(1,len(date)-2):
        for name in names:
            if name not in posi_ret.keys():
                posi_ret[name] = []
            if name not in neg_ret.keys():
                neg_ret[name] = []
            if name not in ret.keys():
                ret[name] = []    
            if daily_begin_capital<=0:
                print(f"在 {date[i]} 破產，結束交易")
                break
            if alpha_030[name][year][i] >=0.125:
                long_pos +=1000 # 假設一次買1000股
                daily_end_capital -= c[name][year][i]* 1000 #以 今天的收盤價 買進(買不到過去的股票)
                buy_price = c[name][year][i]
                print(f"在 {date[i]} 根據策略，以 {buy_price}買進 {name} 股票")
                if daily_end_capital <100000:
                    print('注意：在買進之後之後，資本小於10%')
                sell_price = c[name][year][i+holding_period]
                daily_ret = (sell_price - buy_price)/buy_price
                if daily_ret >=0:
                    posi_ret[name].append(daily_ret)
                else:
                    neg_ret[name].append(daily_ret)
                ret[name].append(daily_ret)
                print(f"在 {date[i+holding_period]} 根據策略，以 {sell_price}賣出 {name} 股票，單次交易報酬：{daily_ret}")
                daily_end_capital += sell_price * 1000
        daily_begin_capital = daily_end_capital
    total_ret = daily_end_capital/initial_capital
    return posi_ret,neg_ret,total_ret
```

# Using Kelly formula to decide the amount of betting

https://bjlkeng.github.io/posts/optimal-betting-and-the-kelly-criterion/ <br>

https://en.wikipedia.org/wiki/Kelly_criterion

$f^{*} = \frac{p}{a}- \frac{q}{b}$

p : 勝率 <br>
q = 1-p <br>
a = the fraction that is lost in a negative outcome <br>
b = the fraction that is gained in a positive outcome

參數p 參考alpha_030_S&P500_Methodology檔案提及之 P(increase | alpha030 >= 0.125) <br>
q = 1-p

a = 策略交易平均虧損(只看虧損)

b = 策略交易平均獲利(只看獲利)

動態調整下注比例： 先假設第一次交易的獲益率為前一年平均獲益率，根據歷史的機率之後計算出第一次下注的比例。 之後得到第一次交易的實際報酬，調整後的發生機率，計算出下一次的下注比例。

```python
p_de = 125 #參考前一份檔案之sum(multiply['AAPL']['2021'] == 1)
p_nu = 67  #參考前一份檔案之intersect_alpha_030
```

```python
p['AAPL'] = p_nu['AAPL']/p_de['AAPL']
p['AMZN'] = p_nu['AMZN']/p_de['AMZN']
q['AAPL'] = 1-p['AAPL']
q['AMZN'] = 1-p['AMZN']
a['AAPL'] = abs(sum(ans[1]['AAPL'])/len(ans[1]['AAPL']))
a['AMZN'] = abs(sum(ans[1]['AMZN'])/len(ans[1]['AMZN']))
b['AAPL'] = sum(ans[0]['AAPL'])/len(ans[0]['AAPL'])
b['AMZN'] = sum(ans[0]['AMZN'])/len(ans[0]['AMZN'])
f['AAPL'] = p['AAPL']/a['AAPL'] - q['AAPL']/b['AAPL']
f['AMZN'] = p['AMZN']/a['AMZN'] - q['AMZN']/b['AMZN']
```

```python
def strategy_1_test(names,year,holding_period):
    """
    name : 股票代碼
    year : 用字串
    """
    date = c[names[0]][year].index.format(formatter=lambda x: x.strftime('%Y %m %d %H:%M.%S'))
    initial_capital = 1000000
    daily_end_capital = initial_capital
    daily_begin_capital = initial_capital
    long_pos = 0
    buy_price = 0
    sell_price = 0
    daily_ret = 0
    earn = []
    ret = {}
    posi_ret_2 = {}
    neg_ret_2 = {}
    var = {}
    risk_exposure = {}
    for i in range(1,len(date)-2):
        for name in names:
            if name not in posi_ret_2.keys():
                posi_ret_2[name] = []
            if name not in neg_ret_2.keys():
                neg_ret_2[name] = []
            if name not in ret.keys():
                ret[name] = []
            if name not in var.keys():
                var[name] = []
            if name not in risk_exposure.keys():
                risk_exposure[name] = []
            if daily_begin_capital<=0:
                print(f"在 {date[i]} 破產，結束交易")
                break
            if alpha_030[name][year][i] >=0.125:
                buy_price = c[name][year][i]
                long_pos = round(f[name]*(1/100) * daily_begin_capital/buy_price) # 假設一次買1000股
                daily_end_capital -= c[name][year][i]* long_pos #以 今天的收盤價 買進(買不到過去的股票)
                print(f"在 {date[i]} 根據策略，以 {buy_price}買進 {name} 股票")
                if daily_end_capital <100000:
                    print('注意：在買進之後之後，資本小於10%')
                p_de[name] +=1
                sell_price = c[name][year][i+holding_period]
                daily_ret = (sell_price - buy_price)/buy_price
                if daily_ret >=0:
                    p_nu[name]+=1
                    posi_ret_pre[name].append(daily_ret)
                    posi_ret_pre[name].pop(0)
                    posi_ret_2[name].append((date[i],daily_ret))
                else:
                    neg_ret_pre[name].append(daily_ret)
                    neg_ret_pre[name].pop(0)
                    neg_ret_2[name].append((date[i],daily_ret))
                ret[name].append(daily_ret)
                print(f"在 {date[i+holding_period]} 根據策略，以 {sell_price}賣出 {name} 股票，單次交易報酬：{daily_ret}")
                daily_end_capital += sell_price * long_pos
                a[name] = abs(sum(neg_ret_pre[name])/len(neg_ret_pre[name]))
                b[name] = sum(posi_ret_pre[name])/len(posi_ret_pre[name])
                p[name] = p_nu[name]/p_de[name]
                f[name] = p[name]/a[name] - q[name]/b[name]
                LGD = sum(neg_ret_pre[name])/len(neg_ret_pre[name])
                risk = q[name] * LGD
                risk_exposure[name].append((date[i],risk))
                value = np.percentile(np.array(ret[name]),5)
                var[name].append(value)
                if daily_ret < value:
                    print(f"在 {date[i+1]} 買賣{name} 股票，報酬率小於VaR")
        daily_begin_capital = daily_end_capital
    total_ret = daily_end_capital/initial_capital
    return posi_ret_2,neg_ret_2,total_ret,var,risk_exposure,ret
```
